```python
import pandas as pd
```

Imports the pandas library and assigns it the alias pd for easy data manipulation and analysis.

```python

data = {
    'age': [25, 30, 22, 40, 55, 60, 33, 28, 45, 50],
    'income': [50000, 60000, 45000, 70000, 80000, 90000, 65000, 62000, 75000, 85000],
    'age_group': ['20-30', '30-40', '20-30', '40-50', '50-60', '50-60', '30-40', '20-30', '40-50', '50-60']
}
df = pd.DataFrame(data)
```

Creates a dictionary containing age, income, and age group data and converts it into a pandas DataFrame named df.

```python
summary_stats = df.groupby('age_group')['income'].describe()
print(summary_stats)
```

Groups the data by age_group and calculates descriptive statistics (count, mean, std, min, max, and quartiles) for the income column.

```python
median_income = df.groupby('age_group')['income'].median()
print("Median Income by Age Group:")
print(median_income)
```

Groups income values by age group and computes the median income for each group.

```python
print("Column Names:", df.columns)
```

Displays the names of all columns present in the DataFrame.

```python

data = {
    'age': [25, 30, 25, 40, 55, 60, 33, 28, 45, 50, 25, 30, 28, 30, 25],
    'income': [50000, 60000, 50000, 70000, 80000, 90000, 65000, 62000, 75000, 85000, 50000, 60000, 62000, 70000, 75000],
    'age_group': ['20-30', '30-40', '20-30', '40-50', '50-60', '50-60', '30-40', '20-30', '40-50', '50-60', '20-30', '30-40', '20-30', '30-40', '20-30']
}
df = pd.DataFrame(data)
```

Creates a new DataFrame with repeated values to allow meaningful calculation of the mode.

```python

mode_age = df['age'].mode()
mode_income = df['income'].mode()
print(f"Mode of Age: {mode_age.values}")
print(f"Mode of Income: {mode_income.values}")
```

Calculates and displays the most frequently occurring values (mode) of age and income.

```python

df = pd.read_csv('iris.csv')
print(df.head()) 
```

Reads the iris.csv file into a DataFrame and displays the first five rows of the dataset.

```python
summary_stats_species = df.groupby('variety').describe()
percentiles = df.groupby('variety').quantile([0.25, 0.5, 0.75])
summary_stats_species = df.groupby('variety').describe()

print("\nPercentiles by Species:")
print(percentiles)
```

Groups the iris dataset by species (variety) and calculates the 25th, 50th (median), and 75th percentiles for numerical features.

```python
summary_stats_species = df.groupby('variety')['sepal.width'].describe()

print("\nSummary Statistics by Species for Sepal Width:")
print(summary_stats_species)
```

Computes descriptive statistics of sepal width for each iris species.

```python
median_values = df.groupby('variety').median()

print("Median Values by Species:")
print(median_values)
```

Calculates the median of all numerical attributes for each iris species.

```python
median_sepal_length = df.groupby('variety')['sepal.length'].median()
print("Median Sepal Length by Species:")
print(median_sepal_length)
```

Finds the median sepal length for each iris species separately.

```python
mode_width = df['sepal.width'].mode()
print(f"Mode of Width: {mode_width.values}")
```

Calculates and displays the most frequently occurring sepal width value in the dataset.
